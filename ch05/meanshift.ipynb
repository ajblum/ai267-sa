```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

### No Global Warming Temperature Drift

```python
# 1990s Temperatures (Baseline)
train = [12, 15, 14, 18, 13, 16, 14, 15, 12, 14]

# 2026 Temperatures (Subtle Shift)
live  = [13, 16, 15, 19, 14, 17, 15, 16, 13, 15]

mu_train, mu_live = np.mean(train), np.mean(live)
print(f"1990 Avg: {mu_train}°C | 2026 Avg: {mu_live}°C")
```

```python
# The T-Test compares the two lists
t_stat, p_val = stats.ttest_ind(train, live)

print(f"P-Value Score: {p_val:.4f}")
print("-" * 30)

# THE RULE: Is the P-Value greater than our 0.05 safety line?
if p_val > 0.05:
    print("RESULT: ✅ NO DRIFT (P-Value is high, change is likely noise)")
else:
    print("RESULT: 🚨 DRIFT DETECTED (P-Value is low, change is real)")
```

```python
plt.figure(figsize=(10, 5))

# Draw the smooth curves
sns.kdeplot(train, fill=True, color="blue", label="1990s (Baseline)")
sns.kdeplot(live, fill=True, color="red", label="2026 (Live)")

# Draw the "Centers" (Means)
plt.axvline(mu_train, color='blue', linestyle='--', label='90s Average')
plt.axvline(mu_live, color='red', linestyle='--', label='26 Average')

plt.title(f"TrustyAI Analysis (P-Value: {p_val:.4f})")
plt.xlabel("Temperature (°C)")
plt.legend()
plt.show()
```

### Global Warming Temperature Drift

```python
# 1990s Temperatures (Same Baseline)
train = [12, 15, 14, 18, 13, 16, 14, 15, 12, 14]

# 2026 Temperatures (Significant 5-Degree Shift)
# We add +5 to every day to simulate a major change
drifted_live = [17, 20, 19, 23, 18, 21, 19, 20, 17, 19]

mu_train, mu_drifted_live = np.mean(train), np.mean(drifted_live)
print(f"1990 Avg: {mu_train}°C | 2026 Avg: {mu_drifted_live}°C")
```

```python
# The T-Test compares the two lists
t_stat, p_val = stats.ttest_ind(train, drifted_live)

print(f"P-Value Score: {p_val:.4f}")
print("-" * 30)

# THE RULE: Is the P-Value greater than our 0.05 safety line?
if p_val > 0.05:
    print("RESULT: ✅ NO DRIFT (P-Value is high, change is likely noise)")
else:
    print("RESULT: 🚨 DRIFT DETECTED (P-Value is low, change is real)")
```

```python
plt.figure(figsize=(10, 5))

# Draw the smooth curves
sns.kdeplot(train, fill=True, color="blue", label="1990s (Baseline)")
sns.kdeplot(drifted_live, fill=True, color="red", label="2026 (DRIFTED)")

# Draw the "Centers" (Means)
plt.axvline(mu_train, color='blue', linestyle='--', label='90s Average')
plt.axvline(mu_drifted_live, color='red', linestyle='--', label='26 Average')

# Annotate the Verdict
plt.title(f"TrustyAI Analysis (P-Value: {p_val:.6f})")
plt.xlabel("Temperature (°C)")

# Logic: Since P-Value is way below 0.05, it's Drift


plt.legend()
plt.show()
```
