```python
from lm_eval import simple_evaluate

# 1. Run multi-task evaluation
# FLAN-T5-Small is an 'Instruct' model, so it understands the 'Question/Answer' format.
results = simple_evaluate(
    model="hf", 
    model_args="pretrained=google/flan-t5-small", 
    tasks=["arc_easy", "hellaswag"], # Passing a list of tasks 
    device="cpu",                    # Running on your workbench CPU [cite: 49]
    limit=10                         # Small limit for a quick live demo [cite: 52]
)

# 2. Extract results dynamically
print(f"{'Task Name':<15} | {'Accuracy':<10}")
print("-" * 30)

for task, metrics in results['results'].items():
    # Dynamically grab the accuracy to avoid KeyErrors [cite: 55]
    acc = metrics.get('acc,none', metrics.get('acc_norm,none', 'N/A'))
    print(f"{task:<15} | {acc:.2%}" if isinstance(acc, float) else f"{task:<15} | {acc}")
```
