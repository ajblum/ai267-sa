```python
import onnxruntime as ort
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
```

```python
# 1. Load Data
X_test = pd.read_csv("X_test_scaled.csv").values.astype(np.float32)
y_test = pd.read_csv("y_test.csv").values
```

```python
# 2. Load ONNX Session
session = ort.InferenceSession("model.onnx")
input_name = session.get_inputs()[0].name
```

```python
# 3. Run Inference
onnx_inputs = {input_name: X_test}
onnx_preds = session.run(None, onnx_inputs)[0]
predictions = (onnx_preds > 0.5).astype(int)
```

```python
# 4. Evaluate
accuracy = accuracy_score(y_test, predictions)
print(f"ONNX Model Accuracy: {accuracy:.4f}")

with open("results.txt", "w") as f:
    f.write(f"ONNX Inference Accuracy: {accuracy}")
```
