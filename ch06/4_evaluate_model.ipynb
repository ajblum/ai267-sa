```python
import onnxruntime as ort
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
```

```python
# 1. Load Data
X_test = pd.read_csv("X_test_scaled.csv").values.astype(np.float32)
y_test = pd.read_csv("y_test.csv").values
```

```python
# 2. Load ONNX Session
session = ort.InferenceSession("model.onnx")
```

```python
# 3. Run Inference
# Use the named input from the export step
onnx_preds = session.run(['output'], {'input': X_test})[0]
predictions = (onnx_preds > 0.5).astype(int)
```

```python
# 4. Evaluate
accuracy = accuracy_score(y_test, predictions)
print(f"Final Model Accuracy: {accuracy:.4f}")
```

```python
# Save results to a text file for S3 persistence
with open("results.txt", "w") as f:
    f.write(f"Diabetes Prediction Model Accuracy: {accuracy}")
```
