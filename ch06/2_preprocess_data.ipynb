```python
import pandas as pd
import numpy as np
```

```python
# Load the artifact from the previous node
df = pd.read_csv("raw_data.csv")
```

```python
# 1. TECHNICAL TIP: Force conversion to numeric. 
# 'coerce' will turn any remaining strings (like accidental headers) into NaN
for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
# Replace 0 with NaN in specific columns where 0 is invalid
cols_to_fix = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
df[cols_to_fix] = df[cols_to_fix].replace(0, np.nan)

# Simple imputation: fill NaN with median
df.fillna(df.median(), inplace=True)
```

```python
# Save cleaned artifact
df.to_csv("cleaned_data.csv", index=False)
print("Data preprocessing complete and types validated.")
```
