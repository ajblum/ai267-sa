```python
import pandas as pd
import os
```

```python
# Pima Indians Diabetes dataset URL copied from https://www.kaggle.com/datasets/uciml/pima-indians-diabetes-database
url = "https://raw.githubusercontent.com/ajblum/ai267-sa/refs/heads/main/ch06/diabetes.csv"
column_names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']

print("Ingesting data from source...")
df = pd.read_csv(url, names=column_names)
```

```python
# Save as output artifact for the next node
df.to_csv("raw_data.csv", index=False)
print("Data saved to raw_data.csv")
```
